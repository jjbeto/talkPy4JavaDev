```python
source_url = 'https://raw.githubusercontent.com/jjbeto/talkPy4JavaDevResources/master/johns_hopkins'
confirmed_file = 'time_series_covid19_confirmed_global.csv'
deaths_file = 'time_series_covid19_deaths_global.csv'
recovered_file = 'time_series_covid19_recovered_global.csv'
target_server = 'http://localhost:8080/api/'
```

```python
import pandas as pd

def load_file(file):
    df = pd.read_csv(f'{source_url}/{file}')
    return df.set_index('country')
```

```python
confirmed = load_file(confirmed_file)
confirmed.head()
```

```python
deaths = load_file(deaths_file)
deaths.head()
```

```python
recovered = load_file(recovered_file)
recovered.head()
```

```python
sick = confirmed - recovered - deaths
sick.head()
```

```python
import numpy as np

pct_change = confirmed.transpose().pct_change()
pct_change = pct_change.replace(np.inf, 1)
pct_change.head()
# print(pct_change['AF'].head(10))
# print(confirmed.loc['AF'].head(10))
```

```python
import math
from datetime import datetime

country_list = []
for country in confirmed.index:
    # { "country": "BR", "values": { "2020-07-10": {"value": 5, "pctChange": 0.25} }}
    country_stat = { "country" : country, "values": {} }
    for date in list(confirmed.columns.values):
        formatted_date = datetime.strptime(date, '%m/%d/%y').strftime('%Y-%m-%d')
        value = confirmed[date][country]
        value_changed = pct_change[country][date]

        country_stat['values'][formatted_date] = {
            "value": int(value),
            "pctChange": float(value_changed) if not math.isnan(value_changed) else None
        }
    country_list.append(country_stat)

country_list
```

```python

def convert_to_statistics(df):
    # calculate pct change
    pct_change = df.transpose().pct_change()
    pct_change = pct_change.replace(np.inf, 1)
    
    country_list = []
    for country in df.index:
        # { "country": "BR", "values": { "2020-07-10": {"value": 0, "pctChange": 0.25} }}
        country_stat = { "country" : country, "values": {} }
        for date in list(df.columns.values):
            formatted_date = datetime.strptime(date, '%m/%d/%y').strftime('%Y-%m-%d')
            value = df[date][country]
            value_changed = pct_change[country][date]

            country_stat['values'][formatted_date] = {
                "value": int(value),
                "pctChange": float(value_changed) if not math.isnan(value_changed) else None
            }
        country_list.append(country_stat)

    return country_list
```

```python

confirmed_json = convert_to_statistics(confirmed)
recovered_json = convert_to_statistics(recovered)
deaths_json = convert_to_statistics(deaths)
sick_json = convert_to_statistics(sick)
```

```python
import requests

url = f'{target_server}confirmed/python'
response = [requests.post(url, json=json).status_code for json in confirmed_json]
{i: response.count(i) for i in response}
```

```python

def send_data(data, target):
    url = f'{target_server}{target}/python'
    response = [requests.post(url, json=json).status_code for json in confirmed_json]
    print(target, {i: response.count(i) for i in response})
```

```python

send_data(confirmed_json, 'confirmed')
send_data(recovered_json, 'recovered')
send_data(deaths_json, 'deaths')
send_data(sick_json, 'sick')
```
